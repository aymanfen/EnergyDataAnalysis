```python
def get_development_status(country):
    developed_countries = [
        'United States', 'Canada', 'United Kingdom', 'Germany', 'France', 'Italy', 
        'Japan', 'Australia', 'New Zealand', 'Austria', 'Belgium', 'Denmark', 
        'Finland', 'Ireland', 'Luxembourg', 'Netherlands', 'Norway', 'Sweden', 
        'Switzerland', 'Iceland', 'Spain', 'Portugal', 'Greece', 'South Korea', 
        'Singapore', 'Israel', 'Slovenia', 'Czechia', 'Estonia', 'Slovakia', 
        'Lithuania', 'Latvia', 'Hungary', 'Poland', 'Croatia', 'Malta', 'Cyprus'
    ]
    
    least_developed_countries = [
        'Afghanistan', 'Angola', 'Bangladesh', 'Benin', 'Bhutan', 'Burkina Faso', 
        'Burundi', 'Cambodia', 'Central African Republic', 'Chad', 'Comoros', 
        'Congo, Democratic Republic of the', 'Djibouti', 'Eritrea', 'Ethiopia', 
        'Gambia', 'Guinea', 'Guinea-Bissau', 'Haiti', 'Kiribati', 'Laos', 'Lesotho', 
        'Liberia', 'Madagascar', 'Malawi', 'Mali', 'Mauritania', 'Mozambique', 
        'Myanmar (Burma)', 'Nepal', 'Niger', 'Rwanda', 'Sao Tome and Principe', 
        'Senegal', 'Sierra Leone', 'Solomon Islands', 'Somalia', 'South Sudan', 
        'Sudan', 'Tanzania', 'Timor-Leste', 'Togo', 'Tuvalu', 'Uganda', 'Vanuatu', 
        'Yemen', 'Zambia'
    ]
    
    if country in developed_countries:
        return 'Developed'
    elif country in least_developed_countries:
        return 'Least Developed'
    else:
        return 'Developing'  # Countries not in either list are classified as developing

# Apply the function to create a new column
df['DevelopmentStatus'] = df['country'].apply(get_development_status)
```
